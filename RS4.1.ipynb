```python
import pandas as pd
```

```python


# Load the dataset
file_path = r"C:\Users\User\Desktop\urbanandruralpopulation new.csv"
df = pd.read_csv(file_path)

column_to_sort = 'Urban population'
ascending_order = True  

# Implementation of Insertion Sort
def insertion_sort(arr, ascending=True):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and (arr[j] > key if ascending else arr[j] < key):
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

column_values = df[column_to_sort].tolist()

# Insertion sort
insertion_sort(column_values, ascending_order)

df[column_to_sort] = column_values
print("Sorting completed and saved to:", column_values)
```
