```python

import pandas as pd
import matplotlib.pyplot as plt

data = {'Country': ['Afghanistan', 'Afghanistan', 'Afghanistan', 'Afghanistan', 'Afghanistan'],
        'Country Code': ['AFG', 'AFG', 'AFG', 'AFG', 'AFG'],
        'Year': [1960, 1961, 1962, 1963, 1964],
        'Urban Population': [724373, 763336, 805062, 849446, 896820],
        'Rural Population': [7898093, 8026804, 8163985, 8308019, 8458694]}

df = pd.DataFrame(data)

df['Total Population'] = df['Urban Population'] + df['Rural Population']

df['Urban %'] = df['Urban Population'] / df['Total Population'] * 100
df['Rural %'] = df['Rural Population'] / df['Total Population'] * 100

# Plot
years = df['Year'].unique()
for year in years:
    df_year = df[df['Year'] == year]
    labels = ['Urban', 'Rural']
    sizes = [df_year['Urban %'].values[0], df_year['Rural %'].values[0]]
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Population Distribution in {year}')
    plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\User\Desktop\urbanandruralpopulation new.csv"
data = pd.read_csv(file_path)

data['Urban population'] = pd.to_numeric(data['Urban population'], errors='coerce')
data['Rural population'] = pd.to_numeric(data['Rural population'], errors='coerce')

country_data = data[data['Code country'] == 'AFG']
urban_population = country_data['Urban population']
rural_population = country_data['Rural population']

plt.figure(figsize=(8, 6))
plt.boxplot([urban_population, rural_population], labels=['Urban population', 'Rural population'])
plt.title('Boxplot of Urban and Rural Population in Afghanistan')
plt.ylabel('Population')
plt.xlabel('Population Type')
plt.grid(True)
plt.show()

```
